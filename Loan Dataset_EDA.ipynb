# Outlier dictection

### *Identify skewed distributions

### *Detect outliers (e.g., extremely overdue loans)

### *Compare spread and central tendency across variables

#### b)clipping

```python
colls=["Principal","terms","age","past_due_days"]
for col in colls:
    plt.figure(figsize=(4,5))
    plt.boxplot(loan_df[col])
    plt.title(f"Boxplot of {col}")
    plt.ylabel(col)
    plt.show()

colls = ["Principal","age","past_due_days"]
for col in colls:
    Q1=loan_df[col].quantile(0.25)
    Q3=loan_df[col].quantile(0.75)
    IQR=Q3-Q1
    lower_limit =Q1-1.5*IQR
    upper_limit =Q3+1.5*IQR
    outlier = loan_df[(loan_df[col] < lower_limit) | (loan_df[col] > upper_limit)]
    print(f"{col} - Number of outliers: {outlier.shape[0]}")
    loan_df[col] = loan_df[col].clip(lower=lower_limit, upper=upper_limit)
```

```python
for col in colls:
    plt.figure(figsize=(5,4))
    plt.boxplot(x=loan_df[col])
    plt.title(f"Boxplot of {col}")
    plt.show()
    
```

### Correlation values range from -1 to 1:

#### +1: perfect positive correlation

```python
figbi, axesbi = plt.subplots(2, 2, figsize=(16, 5))
loan_df.groupby('loan_status')['Principal'].mean().plot(kind='barh',ax=axesbi[0,0],color='skyblue')
loan_df.groupby('loan_status')['terms'].mean().plot(kind='barh',ax=axesbi[0,1],color='salmon')
loan_df.groupby('loan_status')['past_due_days'].mean().plot(kind='barh',ax=axesbi[1,0],color='lightgreen')
loan_df.groupby('loan_status')['age'].mean().plot(kind='barh',ax=axesbi[1,1], color='orchid')
plt.show()
```

##### The 2x2 bar chart visualization provides insight into how key borrower characteristics relate to loan outcomes. If approved loans are associated with higher average principal and longer terms, it suggests that trusted borrowers are offered better loan conditions. A lower average of past due days in the approved group may indicate more responsible repayment behavior. Differences in average age could imply that certain age groups are more likely to receive loan approval.

##### Overall, this visualization helps identify patterns and potential predictors of loan status, such as loan amount, repayment behavior, and borrower demographics.
