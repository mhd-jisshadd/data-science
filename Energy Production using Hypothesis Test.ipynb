```python
num_cols = ['Start_Hour', 'End_Hour', 'Day_of_Year', 'Production']

for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    
    print(f"{col}: {len(outliers)} outliers")
for col in num_cols:
    plt.figure(figsize=(6,4))
    sns.boxplot(x=df[col], color='lightblue')
    plt.title(f'Boxplot of {col}')
    plt.show()
```

">
    Z-TEST
</div>

To compare a sample mean with a population mean.

To compare means of two independent samples.

To test hypotheses about population proportions.

To determine if observed differences are statistically significant.

To help accept or reject the null hypothesis in hypothesis testing.

```python

group1 = df[df['Source'] == 'Solar']['Production'].dropna()
group2 = df[df['Source'] == 'Wind']['Production'].dropna()

z_stat, p_value = ztest(group1, group2, value=0)

print("Z-statistic:", z_stat)
print("P-value:", p_value)
```

```python
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: Significant difference exists.")
else:
    print("Fail to reject the null hypothesis: No significant difference.")
```

Hâ‚€ (Null Hypothesis):
There is no difference in mean production between Solar and Wind.

Hâ‚ (Alternative Hypothesis):
There is a difference in mean production between Solar and Wind.

    T-TEST
</div>

To compare a sample mean with a population mean when the population standard deviation is unknown.

To compare means of two independent samples (independent t-test).

To compare means of two related samples (paired t-test).

To determine if observed differences are statistically significant.

To help accept or reject the null hypothesis in hypothesis testing.

```python
df['Day_Group'] = pd.qcut(df['Day_of_Year'], q=2, labels=['Low', 'High'])

high_prod = df[df['Day_Group'] == 'High']['Production'].dropna()
low_prod = df[df['Day_Group'] == 'Low']['Production'].dropna()

t_stat, p_value = ttest_ind(high_prod, low_prod, equal_var=False)

print("T-statistic:", t_stat)
print("P-value:", p_value)
```

```python
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: Significant difference exists.")
else:
    print("Fail to reject the null hypothesis: No significant difference.")
```

Hâ‚€ (Null Hypothesis):
There is no difference in mean production between High and Low day groups.

Hâ‚ (Alternative Hypothesis):
There is a difference in mean production between High and Low day groups.

">
    CHI2 TEST
</div>

To test the association between two categorical variables (Chi-Square Test of Independence).

To test whether observed frequencies match expected frequencies (Chi-Square Goodness of Fit).

To determine if differences in categorical data are statistically significant.

To help accept or reject the null hypothesis in hypothesis testing.

Useful for count or frequency data, not for means or proportions.

```python
df['Day_Group'] = pd.qcut(df['Day_of_Year'], q=2,)
contingency_table = pd.crosstab(df['Day_Group'], df['Source'])
print("Contingency Table:\n", contingency_table)
chi2, p_value, dof, expected = chi2_contingency(contingency_table)
```

```python
print("\nChi-square Statistic:", chi2)
print("P-value:", p_value)
```

```python
if p_value < 0.05:
    print("There is a statistically significant association between Vehicle class and Day_Group.")
else:
    print("There is no statistically significant association between Vehicle class and Day_Group.")

print("\n")
```

Hâ‚€ (Null Hypothesis):
Day_Group and Source are independent â€” there is no association between the time-of-year group and the energy source type.

Hâ‚ (Alternative Hypothesis):
Day_Group and Source are not independent â€” there is an association between the time-of-year group and the energy source type.

    Anova Test
</div>

To test the association between two categorical variables (Chi-Square Test of Independence).

To test whether observed frequencies match expected frequencies (Chi-Square Goodness of Fit).

To determine if differences in categorical data are statistically significant.

To help accept or reject the null hypothesis in hypothesis testing.

Useful for count or frequency data, not for means or proportions.

```python
anova = df[['Day_of_Year', 'Production']]
anova
```

```python
groups = anova.groupby('Day_of_Year')['Production'].apply(list)
groups
```

```python
f_statistic, p_value = stats.f_oneway(*groups)

print(f"F-statistic: {f_statistic}")
print(f"P-value: {p_value}")
```

```python
if p_value < 0.05:
    print("There are significant differences in Energy Efficiency across Day_of_Year.")
else:
    print("There are no significant differences in Energy Efficiency across Day_of_Year.")
```

">
    F-Test
</div>


To compare the variances of two or more groups to see if they are equal.

To test whether group means are significantly different in ANOVA (Analysis of Variance).

To determine if observed differences among group variances or means are statistically significant.

To help accept or reject the null hypothesis about equality of variances or means.

Often used when comparing more than two groups.

```python
n = len(y_test)  
p = X_train.shape[1]
```

```python
RSS = np.sum((y_test - y_pred) ** 2)
```

```python
TSS = np.sum((y_test - np.mean(y_test)) ** 2)
```

```python
ESS = TSS - RSS
```

```python
MSE = RSS / (n - p - 1)
```

```python
MSR = ESS / p
```

```python
F_statistic = MSR / MSE
```

```python
p_value = 1 - stats.f.cdf(F_statistic, p, n - p - 1)
```

```python
print('F-statistic:', F_statistic)
print('p-value:', p_value)
```

```python
if p_value < 0.05:
    print('The model is significant at the 5% significance level.')
else:
    print('The model is not significant at the 5% significance level.')
```
